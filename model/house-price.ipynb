## Model Training

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

linear_regression = LinearRegression()
```

```python
from sys import maxsize

def get_optimal_model(dataset):
    X = dataset.iloc[:,:-1]
    y = dataset.iloc[:,-1]
    
    lowest_value_of_rmse = maxsize
    best_model = None
    best_scaler = None
    
    for i in range(1000):
        scaler = StandardScaler()
        X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=i)
        X_train = scaler.fit_transform(X_train)
        X_test = scaler.transform(X_test)
        model = linear_regression.fit(X_train, y_train)
        predictions = model.predict(X_test)
        rmse = get_rmse(y_test, predictions)
        
        if rmse < lowest_value_of_rmse:
            lowest_value_of_rmse = rmse
            best_model = model
            best_scaler = scaler
    
    return best_model, best_scaler, lowest_value_of_rmse

def get_rmse(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    return np.sqrt(mse)
```

```python
model, scaler, rmse = get_optimal_model(housing_data)
```
